```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
z =np.array ([2, 9, 5, 8])
x_index = np.arange(len(x))
width = 0.2
plt.bar(x_index,y, color ='r')
plt.bar(x_index+ width ,z,color ='k')
plt.show()
```
